```python
#等高线图
def f(x,y):
    return (1-x/2+x**5 + y**3)*np.exp(-x**2 - y**2)
    

x = np.linspace(-3,3,100)
y = np.linspace(-3,3,100)

X,Y = np.meshgrid(x,y)
plt.contourf(X,Y,f(X,Y),8,alpha=0.75,cmap=plt.cm.hot)

C = plt.contour(X,Y,f(X,Y),8,colors='black',linewidth=.5)
plt.clabel(C,inline=True,fontsize=10)

plt.xticks(())
plt.yticks(())
plt.show()
```

```python
#3D图

from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)

x = np.arange(-4,4,0.25)
y = np.arange(-4,4,0.25)

X,Y = np.meshgrid(x,y)
R = np.sqrt(X**2 + Y**2)
z = np.sin(R)

ax.plot_surface(X,Y,z,rstride=1,cstride=1,cmap=plt.get_cmap('rainbow'))

ax.contourf(X,Y,z,zdir='z',offset=-2,cmap='rainbow')
ax.set_zlim(-2,2)

plt.show()
#换环境可以转动图像
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

x = np.linspace(0, 10, 100)
y = 1+ np.random.randint(0,10,size=(100,1))*0.1
#print(y)

fig, ax = plt.subplots()
line, = ax.plot(x, y, color='k')
line2, = ax.plot(x, y-0.1, color='k')


#print(type(line))

def update(num, x, y, line):
    line.set_data(x[:num], y[:num])
    line.axes.axis([0, 10, 0, 2])
    line2.set_data(x[:num], (y-0.1)[:num])
    line2.axes.axis([0, 10, 0, 2])
    return line,line2  # 可以同时画两个或多个线，只要在update函数中返回多个线即可


ani = animation.FuncAnimation(fig, update,len(x), fargs=[x, y, line],
                              interval=25, blit=True)

ani.save('test.gif')
plt.show()
```

```python
from matplotlib import pyplot as plt
from matplotlib import animation
import numpy as np
fig, ax = plt.subplots()

x = np.arange(0, 2*np.pi, 0.01)
line, = ax.plot(x, np.sin(x))
plt.show()

def animate(i):
    line.set_ydata(np.sin(x + i/10.0))
    return line,

def init():
    line.set_ydata(np.sin(x))
    return line,
ani = animation.FuncAnimation(fig=fig,
                              func=animate,
                              frames=100,
                              init_func=init,
                              interval=20,
                              blit=False)
ani.save('test1.gif')#保存成动画

#保存成视频 ffmpeg 或者mencoder
#ani.save('basic_animation.mp4', fps=30, extra_args=['-vcodec', 'libx264'])
plt.show()
```
