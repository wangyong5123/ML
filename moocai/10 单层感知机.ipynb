```python
#得到的效果不是非常好；
#不能解决非线性问题
import numpy as np
import matplotlib.pyplot as plt
```

```python

X = np.array([[1,3,3],
             [1,4,3],
             [1,1,1],
             [1,0,2]])

Y = np.array([[1],
             [1],
             [-1],
             [-1]])
#权值初始化，3行1列 取值范围-1 1
W = (np.random.random([3,1]) - 0.5) *2

print(W)
#设置学习率
lr = 0.11
#神经网络输出
O = 0

def update():
    global X,Y,W,lr
    O = np.sign(np.dot(X,W))#shape:(3,1)
    W_C = lr*(X.T.dot(Y-O))/int(X.shape[0])# shape 行的个数 ；求取误差的平均值
    W = W + W_C
```

```python
for i in range(100):
    update()#更新权值
    print(W)#打印当前权值
    print(i)#迭代次数
    
    #O = np.sign(np.dot(X,W))#加入了激活函数，如果y=x ？也能学到很好的效果
    O = np.dot(X,W)#加入了激活函数，如果y=x ？也能学到很好的效果
    
    if(O == Y).all():
        print('finished')#如果实际输出等于期望输出，模型收敛，循环结束
        print('epoch:',i)
        break

#正样本 (3,3)(4,3)
x1 = [3,4]
y1 = [3,3]
#负样本(1,1)(0,2)
x2 = [1,0]
y2 = [1,2]

k = -W[1]/W[2]
d = -W[0]/W[2]
# print('k',k)
# print('d',d)

xdata = (0,5)
plt.figure()

plt.plot(xdata,xdata*k+d,'r')
plt.scatter(x1,y1,c='b')
plt.scatter(x2,y2,c='y')
plt.show()
```
